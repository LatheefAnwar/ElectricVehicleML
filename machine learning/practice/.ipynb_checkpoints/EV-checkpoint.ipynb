```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import make_scorer, cohen_kappa_score
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from xgboost import XGBClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
```

**Getting overview of the data and displaying it**

```python
def datacleaning(filename):
    mappers = ['VIN_(1-10)', 'County', 'City', 'State', 'PostalCode', 'ModelYear',
       'Make', 'Model', 'E.V_Type', 'CAFV', 'ElectricRange', 'BaseMSRP',
       'LegislativeDistrict', 'DOLVehicleID', 'VehicleLocation',
       'ElectricUtility', '2020_CensusTract']
    df = pd.read_csv(filename, header=0, low_memory=False, names=mappers)
    df = df.drop(['LegislativeDistrict', 'VIN_(1-10)','BaseMSRP'], axis=1)
    df = df.drop([138102, 182254, 203466], axis= 0)
    df['Model'] = (df.groupby('Make')['Model']
               .transform(lambda x: x.fillna(x.mode()[0] if not x.mode()
                                             .empty else 'unknown')))
    df['VehicleLocation'] = (df.groupby(['City', 'County'])['VehicleLocation']
                         .transform(lambda x: x.fillna(x.mode()[0] if not x.mode()
                                                       .empty else 'Unknown')))
    df['ElectricRange'] = df['ElectricRange'].fillna(46)
    df.loc[(df['ModelYear']==2008) & (df['ElectricRange']==0), 'ElectricRange'] = 220.0
    df = df[~((df['ModelYear'] == 2013) & (df['E.V_Type'] == 'BEV') & (df['ElectricRange'] == 0))]
    df.loc[(df['ModelYear']==2019) & (df['ElectricRange']==0) & (df['Model']=='SOUL EV'), 'ElectricRange'] = 111.0
    df.loc[(df['ModelYear']==2019) & (df['Model'] == 'EQ FORTWO') & (df['ElectricRange']==0), 'ElectricRange'] = 68.0
    df.loc[(df['ModelYear']==2020) & (df['ElectricRange']==0) & (df['Model'] == 'LEAF'), 'ElectricRange'] = 150.0
    df.loc[(df['ModelYear']==2020) & (df['Model'] == 'HARDTOP' ) & (df['ElectricRange']==0), 'ElectricRange'] = 111.0
    df_filtered = df[(df['ModelYear'] >= 2021)]
    yearly_mean = df_filtered[df_filtered['ElectricRange']>0].groupby('ModelYear')['ElectricRange'].mean()
    df_filtered.loc[:,'ElectricRange'] = df_filtered.apply(
    lambda row: yearly_mean[row['ModelYear']] if row['ElectricRange'] == 0 else row['ElectricRange'],
    axis=1
    )
    df.loc[(df['ModelYear']>= 2021) & (df['ElectricRange'] == 0), 'ElectricRange'] = df_filtered['ElectricRange']  
    return df

def preprocessing(df):
    le = LabelEncoder()
    mappers = ['County', 'City', 'State', 'Make', 'Model',
       'E.V_Type', 'CAFV', 'VehicleLocation',
       'ElectricUtility']
    for mapping in mappers:
        df[mapping] = le.fit_transform(df[mapping])
    return df

# def datasplitting(df):
#     x = df.drop(['E.V_Type','County', 'City', 'State', 'PostalCode',
#              'DOLVehicleID', 'VehicleLocation','ElectricUtility', '2020_CensusTract'], axis=1)
#     y = df['E.V_Type']
#     x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, stratify = y)
#     return x_train, x_test, y_train, y_test

def feature_importance(df):
    x = df.drop(['E.V_Type'], axis=1)
    y = df['E.V_Type']
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, stratify = y)
    rf = RandomForestClassifier()
    rf.fit(x_train,y_train)
    importance = rf.feature_importances_
    feature_importance = pd.DataFrame({
        "Features": x_train.columns,
        "Importance": importance
    }).sort_values(by='Importance', ascending=False)
    return feature_importance

def feature_engineering(df):
    df['YearRange'] = df['ModelYear'] + df['ElectricRange']
    df['YearCAFV'] = df['ModelYear'] * df['CAFV']
    return df
```
